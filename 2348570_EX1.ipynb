```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/data - data.csv')
```

```python
df.head()
```

```python
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
summary_stats = df.describe()
print("Summary Statistics:\n", summary_stats)
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='inferno', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
numerical_attributes = ["yr_built", "sqft_above", "price", "bedrooms", "bathrooms"]
df[numerical_attributes].hist(figsize=(12, 8))
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['bedrooms'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.title('Bar Plot of Bedrooms')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
city_counts = df['city'].value_counts()
top_cities = city_counts.head(10)
other_cities_count = city_counts[10:].sum()
top_cities['Other'] = other_cities_count
top_cities.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Pie Chart of Top Cities')
plt.show()
```

```python
df['yr_built Binned'] = pd.cut(df['yr_built'], bins=4)
bin_means_yr_built = df.groupby('yr_built Binned')['yr_built'].mean()
bin_boundaries_yr_built = bin_means_yr_built.index
```

```python
df['sqft_above Binned'] = pd.cut(df['sqft_above'], bins=3)
bin_means_sqft_above = df.groupby('sqft_above Binned')['sqft_above'].mean()
bin_boundaries_sqft_above = bin_means_sqft_above.index
```

```python
# Create bins for "Year Built" and "Sqft_Above"
bins_year_built = [1950, 1970, 1990, 2010, 2023]  # Define your bin boundaries
bins_sqft_above = [0, 1000, 2000, 3000, 5000]  # Define your bin boundaries

# Apply bin means
df['Year_Built_Bin_Means'] = pd.cut(df['yr_built'], bins_year_built, labels=False)
df['Sqft_Above_Bin_Means'] = pd.cut(df['sqft_above'], bins_sqft_above, labels=False)
```

```python
attributes_to_normalize = ["price", "bedrooms", "bathrooms", "sqft_living"]
```

```python
scaler = StandardScaler()
df[attributes_to_normalize] = scaler.fit_transform(df[attributes_to_normalize])
```

```python
df[attributes_to_normalize].hist(figsize=(12, 8))
plt.show()
```

```python
min_max_scaler = MinMaxScaler()
df[attributes_to_normalize] = min_max_scaler.fit_transform(df[attributes_to_normalize])
```

```python
df[attributes_to_normalize].hist(figsize=(12, 8))
plt.show()
```

```python
numerical_attributes = ['price', 'bedrooms', 'bathrooms', 'sqft_living']

# Create a subplot with two rows and one column (for Min-Max and Z-score values)
fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Plot Min-Max values
df_min_max = df[numerical_attributes]
df_min_max.plot(kind='bar', ax=axes[0], colormap='coolwarm', title='Min-Max Normalization')
axes[0].set_ylabel('Value')

# Plot Z-score values
df_z_score = df[numerical_attributes]
df_z_score.plot(kind='bar', ax=axes[1], colormap='coolwarm', title='Z-Score Normalization')
axes[1].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
